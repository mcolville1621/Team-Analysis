# Value of Baserunning
## Different Metrics- BsR, SB%,RS%, UBR

```python
from pybaseball import batting_stats, team_batting
import pandas as pd
```

### 2015-2022 Metrics

```python
Tot_BR = team_batting(2015,2022, ind=0)[["Team", "G", "Spd", "Pace", "BsR", "UBR", "SB", "CS", "R", "H", "HR", "HBP", "BB"]]
list(team_batting(2015,2022, ind=0).columns)
```

```python
Tot_BR["SB%"] = Tot_BR['SB']/(Tot_BR['SB'] + Tot_BR['CS'])
Tot_BR['RS%'] = (Tot_BR['R'] - Tot_BR['HR'])/ (Tot_BR['H'] + Tot_BR['HBP'] + Tot_BR['BB'] - Tot_BR['HR'])
Tot_BR.drop(columns = ["G", "SB", "CS", "R", "H", "HR", "HBP", "BB"], inplace = True)
#RS does not include PR
Tot_BR.sort_values('RS%', ascending = False)
```

### Top 5 BsR- ARI, CLE, TEX, LAD, ATL
### Top 5 UBR- COL, CHC, ATL, TEX, STL
### Top 5 SB%- CLE, ARI, MIL, NYY, KCR
### Top 5 RS%- BOS, COL, HOU, NYY, TEX

### Different teams rank higher on RS metrics than the SB/advanced metrics ex: HOU, BOS, NYY, COL
### Texas seems to best at both advaced and RS metrics
### Looks as though it depends how you look at baserunning metrics to evaluated

## Overall does not seem that looking at baserunning is a factor into how good a team is
### it might help win a few more games or in playoffs may be an important factor, however does not seem super important

```python
Tot_BR.corr()
```

## Best and Worst Baserunners and their Hitting Stats (given at least 200 AB)
